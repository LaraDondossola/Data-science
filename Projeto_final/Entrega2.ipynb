```python
print("==== INCONSISTÊNCIAS ENCONTRADAS ====\n")

# DF1
print("DF1:")
print("- dist_cidade_km negativas:", (df1['dist_cidade_km'] < 0).sum())
print("- sexo inválido:", df1[~df1['sexo'].isin(['M','F'])]['sexo'].unique())
print("- desistiu inválido:", df1[~df1['desistiu'].isin([0,1])]['desistiu'].unique())

# DF2
print("\nDF2:")
print("- sexo inválido:", df2[~df2['sexo'].isin(['M','F'])]['sexo'].unique())

# DF3
print("\nDF3:")
print("- notas fora do padrão:", df3[(df3['nota_final'] < 0) | (df3['nota_final'] > 10)].shape[0])
print("- faltas negativas:", (df3['numero_falta'] < 0).sum())
print("- resultados estranhos:", df3['resultado'].unique())

# DF4
print("\nDF4:")
print("- semestres inválidos (ex: negativos):", (df4['semestre'] < 0).sum())

# DATAS (DF2 e DF4)
for nome, df in {"df2": df2, "df4": df4}.items():
    for col in df.columns:
        if 'dt_' in col:
            temp = pd.to_datetime(df[col], errors='coerce')
            erros = temp.isnull().sum()
            if erros > 0:
                print(f"- {nome} coluna {col} tem {erros} datas inválidas")
```

```python
print("==== INCONSISTÊNCIAS REAIS ====\n")

# DF1
print("DF1:")
print("- dist_cidade_km negativas:")
print(df1[df1['dist_cidade_km'] < 0][['aluno_id', 'dist_cidade_km']])

print("\n- desistiu inválido (esperado 0 ou 1):")
print(df1[~df1['desistiu'].isin([0,1])][['aluno_id', 'desistiu']])


# DF3
print("\nDF3:")
print("- notas inválidas (<0 ou >10):")
print(df3[(df3['nota_final'] < 0) | (df3['nota_final'] > 10)]
      [['novoCodigo', 'nome_disc', 'nota_final']])

print("\n- faltas negativas:")
print(df3[df3['numero_falta'] < 0]
      [['novoCodigo', 'nome_disc', 'numero_falta']])


# DF4
print("\nDF4:")
print("- semestres inválidos (<0):")
print(df4[df4['semestre'] < 0][['novoCodigo', 'semestre']])


# DATAS
for nome, df in {"df2": df2, "df4": df4}.items():
    for col in df.columns:
        if 'dt_' in col:
            temp = pd.to_datetime(df[col], errors='coerce')
            erros = df[temp.isnull()]

            if not erros.empty:
                print(f"\n{nome.upper()} - Datas inválidas na coluna {col}:")
                print(erros[['novoCodigo', col]])
```

```python
print("==== INCONSISTÊNCIAS REAIS ====\n")

# DF1
print("DF1:")
print("- dist_cidade_km negativas:")
print(df1[df1['dist_cidade_km'] < 0][['aluno_id', 'dist_cidade_km']])

print("\n- desistiu inválido (esperado 0 ou 1):")
print(df1[~df1['desistiu'].isin([0,1])][['aluno_id', 'desistiu']])


# DF3
print("\nDF3:")
print("- notas inválidas (<0 ou >10):")
print(df3[(df3['nota_final'] < 0) | (df3['nota_final'] > 10)]
      [['novoCodigo', 'nome_disc', 'nota_final']])

print("\n- faltas negativas:")
print(df3[df3['numero_falta'] < 0]
      [['novoCodigo', 'nome_disc', 'numero_falta']])


# DF4
print("\nDF4:")
print("- semestres inválidos (<0):")
print(df4[df4['semestre'] < 0][['novoCodigo', 'semestre']])


# DATAS (CORRIGIDO)
for nome, df in {"df2": df2, "df4": df4}.items():
    for col in df.columns:
        if 'dt_' in col:
            original = df[col]
            convertido = pd.to_datetime(original, errors='coerce')

            # 🔥 só inválidos reais (não vazios!)
            erros = df[original.notnull() & convertido.isnull()]

            print(f"\n{nome.upper()} - Coluna {col}:")
            print("Vazios:", original.isnull().sum())
            print("Inválidos reais:", erros.shape[0])

            if not erros.empty:
                print("Exemplos de inválidos:")
                print(erros[['novoCodigo', col]].head(10))
            else:
                print("Nenhum valor inválido ✔")
```
