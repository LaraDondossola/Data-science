Criação de Variáveis

```python
# 1. CRIAR NOVAS VARIÁVEIS ÚTEIS (Idade, Mês/Ano)
print("Criando variáveis de tempo e idade...")

# Converter colunas de data para o formato datetime (tratando erros e formatos mistos)
dadosAlunos['dt_nasc'] = pd.to_datetime(dadosAlunos['dt_nasc'], errors='coerce')
# Cortando a string da data de ingresso para pegar apenas a parte principal 'DD/MM/YYYY HH:MM:SS'
dadosAlunos['dt_ingresso'] = pd.to_datetime(dadosAlunos['dt_ingresso'].astype(str).str[:19], errors='coerce')

# Criar a variável 'idade'
data_atual = pd.to_datetime('today')
dadosAlunos['idade'] = (data_atual - dadosAlunos['dt_nasc']).dt.days // 365

# Criar variáveis de 'ano' e 'mês' de ingresso para capturar sazonalidade
dadosAlunos['ano_ingresso'] = dadosAlunos['dt_ingresso'].dt.year
dadosAlunos['mes_ingresso'] = dadosAlunos['dt_ingresso'].dt.month


# 2. AGRUPAR VARIÁVEIS POUCO REPRESENTATIVAS EM CATEGORIAS AMPLAS
print("Agrupando categorias...")

# A) Agrupando 'nome_saida' (Criação de Status Macro)
def agrupar_saida(saida):
    saida = str(saida).upper()
    if 'SEM FORMA SAÍDA' in saida:
        return 'ATIVO'
    elif 'COLAÇÃO DE GRAU' in saida:
        return 'FORMADO'
    else:
        # Engloba Desistência, Trancamento, Abandono, Troca, etc.
        return 'EVADIDO/OUTROS'

dadosAlunos['status_aluno'] = dadosAlunos['nome_saida'].apply(agrupar_saida)

# B) Agrupando 'nome_ingresso' (Criação de Tipo de Ingresso)
def agrupar_ingresso(ingresso):
    if pd.isna(ingresso): return 'OUTROS'
    ingresso = str(ingresso).upper()
    if 'PROUNI' in ingresso:
        return 'BOLSISTA'
    elif 'REMATRICULA' in ingresso:
        return 'REMATRICULA'
    elif 'SELEÇÃO' in ingresso or 'VESTIBULAR' in ingresso:
        return 'PROCESSO_SELETIVO'
    elif 'TRANSFERÊNCIA' in ingresso or 'TROCA' in ingresso:
        return 'TRANSFERENCIA'
    else:
        return 'OUTROS'

dadosAlunos['tipo_ingresso_agrupado'] = dadosAlunos['nome_ingresso'].apply(agrupar_ingresso)

# C) Agrupando Estado Civil ('S' = Solteiro, 'C' = Casado, e agrupando o resto)
dadosAlunos['estado_civil_agrupado'] = dadosAlunos['estado_civil'].map(
    {'S': 'SOLTEIRO', 'C': 'CASADO'}
).fillna('OUTROS')

# D) Agrupando Nacionalidade
dadosAlunos['nacionalidade_agrupada'] = dadosAlunos['nacionalidade'].apply(
    lambda x: 'BRASILEIRO' if pd.notna(x) and 'BRASILEIR' in str(x).upper() else 'ESTRANGEIRO'
)


# 3. TRANSFORMAR CATEGÓRICAS EM DUMMIES (One-Hot Encoding)
print("Criando variáveis Dummies...")

# Selecionar as colunas categóricas recém-criadas ou limpas
cols_categoricas = [
    'sexo',
    'status_aluno',
    'tipo_ingresso_agrupado',
    'estado_civil_agrupado',
    'nacionalidade_agrupada'
]

# Gerar Dummies (True/False ou 1/0 para cada categoria)
# drop_first=True é uma boa prática para evitar multicolinearidade em modelos de ML
dadosAlunos_features = pd.get_dummies(dadosAlunos, columns=cols_categoricas, drop_first=False)


# BÔNUS: Feature Engineering Cruzada (Agrupando dados de Aprovação)
# É muito útil resumir as notas em uma única linha por aluno para alimentar o modelo.
print("Criando métricas de aprovação agregadas...")

df_notas_agrupadas = dadosAprovacao.groupby('novoCodigo').agg(
    media_geral_notas=('nota_final', 'mean'),
    total_faltas=('numero_falta', 'sum'),
    qtd_disciplinas_cursadas=('i_disciplina', 'count')
).reset_index()

# Mesclando as novas métricas de volta para o dataframe de alunos principal
df_final = dadosAlunos_features.merge(df_notas_agrupadas, on='novoCodigo', how='left')

# Preencher possíveis valores nulos nas notas (para alunos que ainda não têm nota)
df_final['media_geral_notas'] = df_final['media_geral_notas'].fillna(0)
df_final['total_faltas'] = df_final['total_faltas'].fillna(0)
df_final['qtd_disciplinas_cursadas'] = df_final['qtd_disciplinas_cursadas'].fillna(0)

# =====================================================================
# EXPORTANDO O RESULTADO
# =====================================================================
print("Processo concluído! O dataset tem", df_final.shape[1], "colunas agora.")
df_final.to_csv('dados_com_features.csv', index=False)
```

**Abaixo está mais uma alteração feita no dataset "Dados Alunos", já que 708 linhas apresentavam dados nulos em: data_igresso, mês_igresso e ano_igresso**

```python
import pandas as pd
import numpy as np

# Carrega o arquivo CSV
df = pd.read_csv('dadosAlunos_limpo (1).csv')

# 1. Garantir que a coluna cp_semestre seja lida como texto (string) para podermos fatiar os números
df['cp_semestre'] = df['cp_semestre'].astype(str)

# 2. Extrair o Ano (os 4 primeiros caracteres de cp_semestre) e converter para número
# Exemplo: '20231' vira 2023
ano_extraido = df['cp_semestre'].str[:4].astype(float)

# 3. Extrair o Semestre (o último caractere de cp_semestre)
# Se for '1', definimos o mês como 1 (Janeiro). Se for '2', definimos como 7 (Julho)
semestre_extraido = df['cp_semestre'].str[-1]
mes_extraido = np.where(semestre_extraido == '1', 1, 7)

# 4. Preencher o ano_ingresso e mes_ingresso APENAS onde o valor atual for nulo (NaN)
df['ano_ingresso'] = df['ano_ingresso'].fillna(ano_extraido).astype(int)
df['mes_ingresso'] = df['mes_ingresso'].fillna(pd.Series(mes_extraido)).astype(int)

# 5. Para a dt_ingresso (Data completa), criamos uma data padrão usando o ano e mês recuperados
# Exemplo: se o ano for 2023 e o mês for 1, vira '2023-01-01'
data_padrao = df['ano_ingresso'].astype(str) + '-' + df['mes_ingresso'].astype(str).str.zfill(2) + '-01'
df['dt_ingresso'] = df['dt_ingresso'].fillna(data_padrao)

# 6. Verifica se todos os nulos foram resolvidos
print("Quantidade de dados nulos após o tratamento:")
print(df[['dt_ingresso', 'ano_ingresso', 'mes_ingresso']].isnull().sum())

df.head()
df.to_csv('dadosAlunos_corrigido.csv', index=False, sep=';')

print("Arquivo corrigido gerado com sucesso!")
```
